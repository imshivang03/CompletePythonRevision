### Filtering a dataframe

```python
ipl.head(2)
```

```python
# Find all the ipl final winners

ipl[ipl['MatchNumber']=='Final'][['Season','WinningTeam']]
```

```python
# Find all the ipl final winners

mask=ipl['MatchNumber']=='Final'
new_df=ipl[mask]
new_df[['Season','WinningTeam']]
```

```python
# How many matches had super over played?

ipl[ipl['WonBy']=='SuperOver'].shape[0]
```

```python
ipl[ipl['SuperOver']=='Y'].shape[0]
```

```python
# How many matches has csk won in kolkata?
ipl[(ipl['City']=='Kolkata') & (ipl['WinningTeam']=='Chennai Super Kings')].shape[0]
```

```python
# Toss winner is match winner in percentage

(ipl[ipl['WinningTeam']==ipl['TossWinner']].shape[0])/ipl.shape[0]*100
```

```python
# Movies with rating higher than 8 and votes>10000

movies[(movies['imdb_rating']>8) & (movies['imdb_votes']>10000)].shape[0]
```

```python
# Action movies with rating greater than 7.5

movies[(movies['genres'].str.split('|').apply(lambda x:'Action' in x)) & (movies['imdb_rating']>7.5)]
```

```python
# Action movies with rating greater than 7.5

mask1=movies['genres'].str.split('|').apply(lambda x:'Action' in x)
mask2=movies['imdb_rating']>7.5

movies[mask1 & mask2]
```

```python
# Action movies with rating greater than 7.5

mask1=movies['genres'].str.contains('Action')
mask2=movies['imdb_rating']>7.5

movies[mask1 & mask2]
```
